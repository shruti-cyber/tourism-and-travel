**Loading of required dataset**

```python
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)
import seaborn as sns

pd.options.mode.chained_assignment = None



# read the data
df = pd.read_csv('data.csv')

# shape and data types of the data
print(df.shape)
print(df.dtypes)
```

**Count of different Categories (Artist Studio , Gallery , Museum, Perfromance Space , Rehreasal and Studio) of Destinations**

```python
df['category'].value_counts().plot(kind='bar' , color=['black', 'red', 'green', 'blue', 'cyan'])
```

**Count of Places according thier catergories with user ratings = 5**

```python
df_good_rating = df[df['ratings']==5]
df_rating_by_category = df_good_rating.groupby(["category"])["place_name"].count()




```

```python
df_rating_by_category.plot(kind="pie")
```

**Count of Places according thier catergories with user ratings = 1**

```python
df_bad_rating = df[df['ratings']==1]
df_rating_by_category = df_bad_rating.groupby(["category"])["place_name"].count()
df_rating_by_category.plot(kind="pie")
```

**List  different agencies owing places different cities of British Columbia**

```python
df_owned = df.groupby(["city","ownership"])["ownership"].count()
df_owned.head(50)
```

```python
df_owned.head(20).plot()
```

**The count of number of places which were famous amongst tourist in the year = 2020**

```python
df_most_rated_place_2020 =df.loc[(df["year"]==2020) & (df["ratings"]==5),"place_name"].count()
```

```python
df_most_rated_place_2020
```

** The count of number of places visited by tourists yearwise**

```python
df_count_places_visited_year_wise=df.groupby(["year"]) ["place_name"].count()
df_count_places_visited_year_wise
```

```python
df_count_places_visited_year_wise.plot(kind="bar",x="year",y="count of places visited" , color=['black', 'red', 'green', 'blue', 'cyan'])
plt.title("count of places visited yearwise")
plt.xlabel("Year")
plt.ylabel("count")
```

```python
df_citywise = df.groupby(["city", "year"]) ["place_name"].count()
df_citywise
```

```python
df_citywise.head(20).plot(kind="bar", color=['black', 'red', 'green', 'blue', 'cyan'])
plt.title("count of places visited in different cities yearwise")
plt.xlabel("Year & City")
plt.ylabel("count")
```

```python
df_categorywise_count=df.groupby(["category","city"]) ["place_name"].count()
```

```python
df_categorywise_count.tail(50).plot(kind="bar")
plt.title("count of places visited in different cities categorywise")
plt.xlabel("Year & Category")
plt.ylabel("count")
```
